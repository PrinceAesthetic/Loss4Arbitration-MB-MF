```python
import pickle
import pandas as pd

# === 读取文件 ===
with open("PKL_DATA/fitdata/fitresults_ALL_larbi_leaky.pkl", "rb") as f:
    results = pickle.load(f)

# === 提取参数 ===
# leak
records = []
for res in results:
    subj = res.get("subj", "unknown")
    names = res.get("param_name", [])
    values = res.get("param", [])
    for name, val in zip(names, values):
        if "leak" in name.lower():  
            records.append({"subj": subj, "param": name, "value": val})

# === 转为表格 ===
df_leak = pd.DataFrame(records)
print(df_leak.head(10))

# 若想看每个被试的 theta 均值
df_mean = df_leak.groupby("subj")["value"].mean().reset_index(name="mean_leak")
print("\n每个被试的 leak 平均值：")
print(df_mean)

# 保存结果
df_leak.to_csv("leak_larbi_leaky.csv", index=False)
print("\n✅ 结果已保存为 leak_larbi_leaky.csv")

# tau
records = []
for res in results:
    subj = res.get("subj", "unknown")
    names = res.get("param_name", [])
    values = res.get("param", [])
    for name, val in zip(names, values):
        if "tau" in name.lower():  
            records.append({"subj": subj, "param": name, "value": val})

# === 转为表格 ===
df_tau = pd.DataFrame(records)
print(df_tau.head(10))

# 若想看每个被试的 tau 均值
df_mean = df_tau.groupby("subj")["value"].mean().reset_index(name="mean_tau")
print("\n每个被试的 tau 平均值：")
print(df_mean)

# 保存结果
df_tau.to_csv("tau_larbi_leaky.csv", index=False)
print("\n✅ 结果已保存为 tau_larbi_leaky.csv")


# theta
records = []
for res in results:
    subj = res.get("subj", "unknown")
    names = res.get("param_name", [])
    values = res.get("param", [])
    for name, val in zip(names, values):
        if "theta" in name.lower():  
            records.append({"subj": subj, "param": name, "value": val})

# === 转为表格 ===
df_theta = pd.DataFrame(records)
print(df_theta.head(10))
# 若想看每个被试的 theta 均值
df_mean = df_theta.groupby("subj")["value"].mean().reset_index(name="mean_theta")
print("\n每个被试的 theta 平均值：")
print(df_mean)
# 保存结果
df_theta.to_csv("theta_larbi_leaky.csv", index=False)
print("\n✅ 结果已保存为 theta_larbi_leaky.csv")

# alpha_u
records = []
for res in results:
    subj = res.get("subj", "unknown")
    names = res.get("param_name", [])
    values = res.get("param", [])
    for name, val in zip(names, values):
        if "alpha_u" in name.lower():  
            records.append({"subj": subj, "param": name, "value": val})


# === 转为表格 ===
df_alpha_u = pd.DataFrame(records)
print(df_alpha_u.head(10))  
# 若想看每个被试的 alpha_u 均值
df_mean = df_alpha_u.groupby("subj")["value"].mean().reset_index(name="mean_alpha_u")
print("\n每个被试的 alpha_u 平均值：")
print(df_mean)
# 保存结果
df_alpha_u.to_csv("alpha_u_larbi_leaky.csv", index=False)
print("\n✅ 结果已保存为 alpha_u_larbi_leaky.csv")
```
